### missing values

```python
# let's see the missing values in categorical features

categorical_nan_feature = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and data[feature].dtype == 'O']

for features in categorical_nan_feature:
    print(features, np.round(df[features].isnull().mean(), 4), "% missing values")
```

```python
# lets replace the missing value with another label

def replace_missing(df, categorical_nan_feature):
    data = df.copy()
    data[categorical_nan_feature] = data[categorical_nan_feature].fillna("Missing")
    return data

df = replace_missing(df, categorical_nan_feature)

df[categorical_nan_feature].isnull().sum()
```

```python
df.head()
```

```python

numerical_with_nan = [feature for feature in df.columns if df[feature].isnull().sum() > 1 and df[feature].dtypes != 'O']

for feature in numerical_with_nan:
    print("{}: {}% missing value".format(feature,np.around(df[feature].isnull().mean(),4)))
```

```python
# lets replace the nan value in numerical feature with some meaningful value i.e. median

for feature in numerical_with_nan:
    median_val = df[feature].median()

    ## creating new feature to capture nan value -> that would give us some usefull information

    df[feature + 'nan'] = np.where(df[feature].isnull(), 1, 0)
    df[feature].fillna(median_val, inplace = True)

df[numerical_with_nan].isnull().sum()
```

```python
df.head(50)
```

```python
# now lets handle temporal features

for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:

    df[feature] = df['YrSold']  - df[feature]
```

```python
df[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
# since numerical feature are skewed, we have to perform lognormal tarnsformation

num_features = ['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_features:
    df[feature] = np.log(df[feature])
```

# Handling Rare Categorical Feature
We will remove categorical variables that are present less than 1% of the observations

```python
categorical_features = [feature for feature in df.columns if df[feature].dtype=='O']
```

```python
categorical_feature
```

```python
for feature in categorical_features:
    temp = df.groupby(feature)['SalePrice'].count()/len(df)
    temp_df = temp[temp > 0.01].index
    df[feature] = np.where(df[feature].isin(temp_df), df[feature], 'Rare_var')
```

```python
df.head(50)
```

```python
# This code is doing ordinal encoding based on target mean.It converts categorical values into numbers according to the average SalePrice of each category.

for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}    # dictionary comprehension
    df[feature]=df[feature].map(labels_ordered)
```

# Feature Selection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

pd.pandas.set_option("display.max_columns", None)
```

```python
dataset = pd.read_csv("X_train.csv")
```

```python
dataset.head()
```

```python
y_train = dataset[['SalePrice']]
```

```python
X_train = dataset.drop(['Id', 'SalePrice'], axis = 1)
```

```python
### Apply Feature Selection
# first, I specify the Lasso Regression model, and I
# select a suitable alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected.

# Then I use the selectFromModel object from sklearn, which
# will select the features which coefficients are non-zero

feature_sel_model = SelectFromModel(Lasso(alpha = 0.005, random_state = 0))
feature_sel_model.fit(X_train, y_train)
```

```python
feature_sel_model.get_support()        # True -> that feature important, False -> skip the feature
```

```python
# let's print the number of total and selected features

# this is how we can make a list of the selected features
selected_feat = X_train.columns[feature_sel_model.get_support()]

print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(feature_sel_model.estimator_.coef_ == 0)))

```
